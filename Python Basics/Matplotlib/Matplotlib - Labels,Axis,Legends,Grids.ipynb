```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plt.plot([1],[1],"rD")
days=[1,2,3,4,5,6,7]
max_t=[41,51,62,32,42,62,42]
min_t=[26,11,32,35,52,32,22]
avg_t=[30,51,27,53,45,32,52]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("WEATHER")
plt.plot(days,max_t,color="red", linestyle="--",label="Max")
plt.plot(days,min_t,color="green", linestyle="--",label="Min")
plt.plot(days,avg_t,color="blue", linestyle="--",label="Avg")

plt.legend(loc="best",shadow=True,fontsize="large")
plt.grid()
```
