```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113,124,412,115,435,56,247,17,351,567]
plt.hist([blood_sugar],bins=8,rwidth=.9)
```

```python
plt.hist([blood_sugar],bins=[110,210,310,410,510,610],rwidth=.9, color='g')
```

```python
plt.hist([blood_sugar],bins=[110,210,310,410,510,610],rwidth=.9, color='g',histtype='step')
```

```python
blood_sugar_1 = [113,124,412,115,435,56,247,17,351,567]
blood_sugar_2 = [113,124,412,115,435,56,247,17,351,567]
plt.hist([blood_sugar_1,blood_sugar_2],
         color=["green","blue"],
         rwidth=.9,
         bins=[50,100,150,200,250,300,350,400,450,500],
         label=["Men","Women"]
        )
plt.title("Blood Sugar Analysis")
plt.xlabel("Suger Level")
plt.ylabel("Frequency Patients")
# if you use labels, you have to enable legend()
plt.legend()
```

```python
blood_sugar_1 = [113,124,412,115,435,56,247,17,351,567]
blood_sugar_2 = [113,124,412,115,435,56,247,17,351,567]
plt.hist([blood_sugar_1,blood_sugar_2],
         color=["green","blue"],
         rwidth=.9,
         bins=[50,100,150,200,250,300,350,400,450,500],
         label=["Men","Women"],
         orientation="horizontal"
        )
plt.title("Blood Sugar Analysis")
plt.xlabel("Suger Level")
plt.ylabel("Frequency Patients")
# if you use labels, you have to enable legend()
plt.legend()
```
