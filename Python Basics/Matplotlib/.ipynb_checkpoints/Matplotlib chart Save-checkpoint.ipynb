```python
import pandas as pd;
import numpy as np
import matplotlib.pyplot as plt
val = np.array([123,421,556,213])
val_labels=["A","B","C","D"]
plt.pie(val,labels=val_labels,autopct="%1.f%%",shadow=False,explode=[0.03,0.03,0.03,0.03], startangle=180)
plt.axis("equal")
plt.show()
plt.savefig("asds.png", dpi=None, facecolor='w', edgecolor='w', orientation='portrait', transparent=False, bbox_inches=None )




```
