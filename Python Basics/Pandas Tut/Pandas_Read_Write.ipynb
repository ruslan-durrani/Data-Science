```python
df = pd.read_csv("stock_data.csv", header = 0, names=["tickers","eps","revenue","price","people"], na_values={
    'eps':["not available","n.a."],
    'price':["not available","n.a."],
    'people':["n.a.","0"]
})
df
```

```python
df.columns
```

```python
df.to_csv("NEW.csv",header=None)
pd.read_csv("NEW.csv",names=["tickers","EPS","Revencue","price","People"])
```

```python
def converter_func(cell):
    if cell == "n.a.":
        return "Ruslan"
    return cell
def eps_converter(cell):
    if cell == "not available":
        return 0
    return cell
df = pd.read_excel("stock_data.xlsx",header=0, converters={
    "people":converter_func,
    "eps":eps_converter,
})
df
```

```python
df.to_excel("NEW.xlsx",sheet_name="Stocks_Sheet",startrow=0,startcol=0, index=False)
pd.read_excel("NEW.xlsx", skiprows=1, header=None, names=["Tickers","Eps","P1","P2","P3"])
```
