```python
import pandas as pd
pakistan_city_temp = pd.DataFrame({
    "city":["ISLAMABAD","RAWALPINDI","LAHORE","QUETA","PESHAWAR","KARACHI", "Pushpapura"],
    "temperature":[32,52,23,42,12,4,2],

})
pakistan_city_humidity = pd.DataFrame(
{
    "city":["ISLAMABAD","RAWALPINDI","LAHORE","QUETA","PESHAWAR","KARACHI","PUushpapura"],
    "humidity":[10,8,13,51,2,14,1],
}
)

# city = pd.Series(["ISLAMABAD","RAWALPINDI","LAHORE","QUETA","PESHAWAR","KARACHI"], name="city")
# temp = pd.Series([32,52,23,42,12,4],name="temperature")
# hum = pd.Series([10,8,13,51,2,14],name="humiduty")
```

```python
df = pd.merge(pakistan_city_temp,pakistan_city_humidity, on="city")
df
```

```python
df = pd.merge(pakistan_city_temp,pakistan_city_humidity, on="city", how = "inner", indicator=True) # intersetion of both DataFrames
df
```

```python
df = pd.merge(pakistan_city_temp,pakistan_city_humidity, on="city", how = "outer", indicator=True) # union of NaN null or 0 
df
```

```python
df = pd.merge(pakistan_city_temp,pakistan_city_humidity, on="city", how = "left", indicator=True) # union of NaN null or 0 
df
```

```python
df = pd.merge(pakistan_city_temp,pakistan_city_humidity, on="city", how = "right", indicator=True) # union of NaN null or 0 
df
```
