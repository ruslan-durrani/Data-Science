```python
df["zscore"] = (df.Height - df.Height.mean())/df.Height.std()
```

```python
# Verify th esolution
u = df.Height.mean()
```

```python
meu = df.Height.std()
```

```python
x = df.Height
```

```python
zscore = (x-u)/meu
zscore.head()
```

```python
df[(df.zscore<-3)|(df.zscore>3)] #Outliers
```
